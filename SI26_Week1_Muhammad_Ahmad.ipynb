1. What is OCR (Optical Character Recognition)?
Ans:

The Optical Character Recognition technology is a wonderful one which can recognize texts in the form of images, scanned documents and photographs and convert them into editable texts on the computer. This shows that through the Optical Character Recognition technology, the computer gets a chance to comprehend the information written in characters, numbers and symbols without requiring manual input.
Optical Character Recognition technology is very widely used in making copies of books, forms and important documents.

2. Why is Urdu OCR harder than English OCR?
Ans:

The process of Urdu Optical Character Recognition is quite difficult in nature. The reason behind this difficulty is that most of the letters of Urdu language are joined together and their shapes vary according to their position in the word.
Dots and diacritics are extremely crucial for Urdu Optical Character Recognition. These dots and diacritics play a vital role in recognizing the characters while implementing Urdu Optical Character Recognition.
It is true that even a minor error in reading these characteristics of Urdu Optical Character Recognition may lead to change in the meaning of the word itself in Urdu Optical Character Recognition.
