Question 1
What is OCR (Optical Character Recognition)?

Answer:

Optical Character Recognition (OCR) is a technology that converts text from images or scanned documents into editable digital text. It allows computers to recognize printed or handwritten characters. OCR saves time by reducing manual typing and makes documents easier to edit and search.

Question 2
Why is Urdu OCR harder than English OCR?

Answer:

Urdu OCR is more difficult because Urdu letters are connected and change shape based on their position in a word. Small dots and marks are also important for identifying characters correctly. These features make Urdu text harder to recognize than English.

Question 3
What are two real-world situations where Urdu OCR would be useful?

Answer:

Urdu OCR can be used to convert old Urdu books and newspapers into digital text for easy storage and searching. It is also useful for digitizing printed forms and official documents in schools, banks, and government offices. This reduces manual work and improves efficiency.
